READING THE DATASET

```python
zip_file_path = 'D:\AI\Data-sets\P_Data_Extract_From_World_Development_Indicators.zip'
file_name = '416432ef-c5ee-4284-b4b6-9f197f6dfdca_Data.csv'

with zipfile.ZipFile(zip_file_path,'r') as zip_ref:
    with zip_ref.open(file_name) as file:
        dc = pd.read_csv(file)
```

DATA CLEANING

```python
dc = dc.drop(index = range(266,271))
dc
```

```python
dc.replace('..', np.nan, inplace = True)
```

```python
dc
```

```python
dc.isnull().sum()
```

```python
dc.info()
```

```python
dc = dc.drop(['Series Name', 'Series Code', 'Country Code', '1990 [YR1990]', '2000 [YR2000]'], axis = 1)
```

```python
dc
```

```python
rows_with_missing_values = dc[dc.isnull().any(axis = 1)]
```

```python
rows_with_missing_values
```

```python
threshold = 7
dc_remove = dc.dropna(thresh = len(rows_with_missing_values.columns) - (threshold + 1))
```

```python
dc_remove
```

```python
country = dc_remove['Country Name']
country.isnull().sum()
```

```python
dc.drop(['Country Name'], axis =1, inplace = True)
dc
```

```python
country
```

```python
dc_filter = dc_remove.T
dc_filter
```

```python
dc_filter = dc_filter.iloc[1:]
dc_filter
```

```python
from sklearn.impute import SimpleImputer
```

```python
values = dc_filter.values
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

```python
transformed_values = imputer.fit_transform(values)
```

```python
dc_update = pd.DataFrame(transformed_values)
dc_update
```

```python
dc_update.isnull().sum()
```

```python
dc_update = dc_update.T
dc_update
```

```python
country_update = country.reset_index(drop = True)
```

```python
dc_update_new = dc_update.reset_index(drop = True)
```

```python
dc_final = pd.concat([country_update, dc_update_new], axis = 1)
dc_final
```

```python
dc_final.isnull().sum()
```
